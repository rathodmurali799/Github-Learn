```python
import pandas as pd
```

```python
df=pd.read_csv("zomato.csv", encoding='latin1')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df=df.iloc[:,1:] # preprocess dataset
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt ## EDA DATA VISUALIZATION
```

```python
plt.scatter(df['City'],df['Currency'])
```

```python
X=df.iloc[:,0:2] # two independable values input and output values
y=df.iloc[:,-1] # dependable valuew
```

```python
X
```

```python
y
```

```python
# scale the values
# trian and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler # scalers the values
```

```python
sc=StandardScaler()
```

```python
X_train = sc.fit_transform(X_train[['Country Code']])
```

```python
X_train
```

```python
X_test = sc.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression ## train the models
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train, y_train.values, clf=clf,legend=2)
```
